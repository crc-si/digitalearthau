This application is to auto generate the docs for all products in the datacube using one template file, **'find_Landsat_metadata.ipynb'**

## 1. Create a Tmp dir to create the files

- Place **find_Landsat_metadata.ipynb** in any directory (BASEDIR) and create a **Tmp** dir in it
- Place this notebook (**replicate.ipynb**) in BASEDIR/TMP.


## 2. Make a list of products in the datacube

- Open the above file in any editor and,
    - Comment out **#dc.list_products()**
    - Uncomment **#custom_list...**
    - Comment out **print ("{}. {} ...** in function **custom_list**
    - Uncomment **print(i['name'])** in function **custom_list**
    - Run the above block of code (after executing all the code above it)
    - Copy the list of products that will be displayed in the box below.

## 3. Copy the template file to make a TXT file for each product
- Create a bash script (e.g. copy.sh) as below...
    - Insert in it...
        - cd BASEDIR/Tmp
    - Insert the following lines (one line per product)
        - cp \-f ../find_Landsat_metadata.ipynb ./find_{product}_metadata.txt
        
## 4. Copy the template to the TXT files in Tmp dir

   - Change dir to BASEDIR/Tmp, where 'copy.sh' resides and execute it 
       -  **source copy.sh**
   - Keep this copy.sh file for later use. Recreate it only if the products change
   - Create a file, '**list_ipnyb.txt**', containing the names of all TXT files, one per line as below
   
       - find_{product}_metadata.txt

## 5. Create the *product.ipynb* files

- Execute this notebook (**replicate.ipynb**) by the **Cell | Run All** command

- It will create an **ipynb** file for each product. 
    - The explanatory text in all will be the same as in the template (**find_Landsat_metadata.ipynb**) but on executing them their respective product details will be retrieved.

- You may optionally remove the TXT files in the BASEDIR/Tmp, but is not necessary.

## 6. Execute all *product.ipynb* files

- Create a 'execute.sh' file with the following lines (one per product file)

    - jupyter nbconvert --execute --to notebook --inplace <find_{product}_metadata.ipynb>

- Execute it by the **source execute.sh** command
    - It will now run all files and rewrite the content to the same filename
    - It is recommended that you open each one in Jupyter notebook to make sure that there is no execution error
- Keep the file for later use.

## Create the index.html

Before creating the HTML files from these **ipynb** files, each of them must be executed and saved again. For this, open them individually in **Jupyter notebook** and run by the **Cell | Run All** and save. There is no need to edit anything, but it is recommended to check that no error has happened on execution. 

There is a way to execute and create HTML files from these **ipynb** files at the commandline, but I am not sure whether it is the desired method. The command below can be listed for each file in a bash script and executed in one go, as in copy.sh and execute.sh described above.

- jupyter nbconvert --execute <find_{product}_metadata.ipynb>

If you need to make changes, edit the template file (**find_Landsat_metadata.ipynb**) and recreate the product files by doing the steps 4 to 6 above. It will save you several hours of manual effort required in creating the files separately.

```python
filename = 'list_ipnyb.txt'
with open(filename) as f:
    content = f.readlines()
    content = [x.strip() for x in content] 
for file in content:
    page_title = file.title()
    page_title = page_title.replace(".Txt","")
    product = file.replace("find_","")
    product = product.replace("_metadata.txt","")
#    outfile = file
    outfile = file.replace(".txt", ".ipynb")
    f2 = open(outfile, "w+")
    print ("Page: ",page_title)
    print ("  Product: ",product)
    with open(file) as f1:
        content1 = f1.readlines()
        for line in content1:
            line = line[:-1]
#            line = line.replace("Find_Landsat7_Data.ipynb", page_title)
            line = line.replace("ls7_nbar_albers", product)
            f2.write(line)
        f2.close()
        break
```
